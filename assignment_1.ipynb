```python
import matplotlib.pyplot as plt

data = [1,7,11,14,17,20,28,34,45,59,63,93,114,147,199,253,306,367,438,495,658,840,1036,1264,1534,1836,2337]
data2 = [data[i] - data[i-1] for i in range(1,len(data))]
```

```python
days = list(range(5,32))
labels = [f'{x}-03' for x in days]
plt.bar(height=data, x=days);
plt.xlabel("Day in March");
plt.ylabel("Total Covid-19 cases");
plt.xticks(days[::5], labels[::5], rotation=45);
plt.title("Total Covid-19 cases in Russia");
plt.savefig("total_cases.jpg")
```

```python
plt.plot(days[1:],data2);
plt.xlabel("Day in March");
plt.ylabel("Total new cases");
plt.xticks(days[::5], labels[::5], rotation=45);
plt.title("Total new cases per day in Russia");
plt.savefig('new_cases_per_day.jpg')
```
